````python
def download_files(url, file_name):
    # Check if the file already exists
    if os.path.exists(file_name):
        print(f"The file {file_name} already exists.")
        return
    
    # now in the directory we can download the files via http
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.'
    }

    req = urllib.request.Request(url, data=None, headers=headers)
    with urllib.request.urlopen(req) as response:
       # Read the response as bytes
       file_data = response.read()

    # Now let's save the file locally
    with open(file_name, 'wb') as f:
       f.write(file_data)

    print(f"{file_name} has been downloaded successfully.")

url = "https://www.py4e.com/code3/words.txt"
url2 = "https://www.py4e.com/code3/romeo.txt"
file_name1 = "words.txt"
file_name2 = "romeo.txt"
download_files(url, file_name1)
download_files(url2, file_name2)
````

Write a Python program that reads the words.txt text file and calculates the frequency (i.e. creates a histogram) of each word in the file. You should use a dictionary to store the word counts, where the keys are the unique words, and the values are the corresponding frequencies. The program should then print the words and their frequencies, sorted in descending order of frequency. Only print words that have a frequency of 2 or more, and exclude words with a count of 1. If there are words with the same frequency, they should be sorted alphabetically.

Download words.txt (used in the chapters) for this exercise.
https://www.py4e.com/code3/words.txt

Sample output (abbreviated):

Words 5
Frequency 3
Program 3
Should 3
File 2
Frequencies 2

````python
def read_file(file_path):
    with open(file_path, 'r') as file:
        return file.read()
    
def word_frequency(text):
    words = text.split()
    frequency = {}
    
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1

    return frequency

def sort_words_by_length(text):
    words = text.split()
    word_lengths = [(word, len(word)) for word in words]
    return sorted(word_lengths, key=lambda x: (-x[1], x[0]))

def filter_and_sort(frequency):
    filtered = {word: count for word, count in frequency.items() if count > 1}
    return sorted(filtered.items(), key=lambda x: (-x[1], x[0]))
````

````python
file_path = 'words.txt'
text = read_file(file_path)
frequency = word_frequency(text)
sorted_words = filter_and_sort(frequency)

for word, count in sorted_words:
    print(f"{word}: {count}")
````

Write a Python function that takes a list of tuples as input, where each tuple contains a student's name and their corresponding grades in different subjects. The function should return a dictionary where the keys are student names and the values are tuples of their average grade (round to one decimal place) and the highest grade they achieved across all subjects.

Example input & output:

```
student_grades = [('Clement', (90, 85, 88)), ('Amy', (85, 90, 92))]
calculate_student_stats(student_grades)
# Output: {'Clement': (87.7, 90), 'Amy': (89.0, 92)}
```

````python
def calculate_student_average_score(student_grades):
    students = {}
    
    for student, grades in student_grades:
        average_grade = round(sum(grades) / len(grades), 1)
        highest_grade = max(grades)
        students[student] = (average_grade, highest_grade)
        
    return students

student_grades = [('Clement', (90, 85, 88)), ('Amy', (85, 90, 92))]
print(calculate_student_average_score(student_grades))
````

Write a Python program that reads the words in romeo.txt and creates a dictionary where each key is the word in all lowercase letters and the value is the same word in all uppercase letters. Each word should only appear once, and words should be sorted in alphabetical order within the dictionary.

Download romeo.txt (used in the chapters) for this exercise.
www.py4e.com/code3/romeo.txt

````python
def create_dictionary(text):
    words = text.split()
    word_dict = {}

    for word in words:
        lowercase_word = word.lower()
        word_dict[lowercase_word] = lowercase_word.upper()

    return dict(sorted(word_dict.items()))
````

````python
data = read_file('romeo.txt')
dictionary = create_dictionary(data)
for key, value in dictionary.items():
        print(f"{key}: {value}")
````

Write a python program that merges multiple tuples into one, and results in a tuple that contains no duplicate elements. 
```
Tup1 = (hail, to, the, conquering, heroes)
Tup2 = (hail, hail, to, Michigan, the, leaders, and, best)
Tup3 = (hail, to, the, victors, valiant) 
```

````python

Tup1 = ("hail", "to", "the", "conquering", "heroes")
Tup2 = ("hail", "hail", "to", "Michigan", "the", "leaders", "and", "best")
Tup3 = ("hail", "to", "the", "victors", "valiant")

def merge_tuples_into_one(*tuples):
    # read a list of tuples and merge the tuples into one tuple
    merged_set = set()
    for tup in tuples:
        merged_set.update(tup)
    return tuple(merged_set)

merged_tuple = merge_tuples_into_one(Tup1, Tup2, Tup3)
print(merged_tuple)
````
